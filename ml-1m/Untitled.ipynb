```python
import pandas as pd
import numpy as np
```

```python
users=pd.read_table("users.dat",sep='::',header=None,names=['UserID','Gender','Age','Occupation','Zipcode'],engine='python')
```

```python
movies=pd.read_table("movies.dat",sep='::',header=None,names=['MovieID','Title','Genre'],engine='python')
movies['Year']=[x[-5:-1] for x in movies.Title]
movies['Title']=[x[:-6] for x in movies.Title]
```

```python
movies.head()
```

```python
Genres=['Action','Adventure','Animation','Children','Comedy','Crime','Documentary','Drama','Fantasy','Film','Horror','Musical','Mystery','Romance','Sci','Thriller','War','Western']
for x in Genres:
    movies[x]=None
```

```python
for movieid in movies.MovieID:
    Genrelist=str(movies.loc[movies.MovieID==movieid,'Genre'])
    for genre in Genres:
        if genre in Genrelist:
            movies.loc[movies.MovieID==movieid,genre]=1
```
